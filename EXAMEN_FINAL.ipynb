```python
import csv
f2015 = open('2015.csv', mode='rt')
f2016= open('2016.csv', mode='rt')
```

```python
import pandas as pd
f2015=pd.read_csv('2015.csv')
f2016=pd.read_csv('2016.csv')
```

```python
f2015
```

```python
f2016

#Se observan diferencias entre las columnas, el conjunto de 2016 presenta 2 columnas nuevas (Lower confidence interval and upper confidence interval) y no presenta la columna del error estandar que se observa en el dataset de 2015.
```

```python
data = pd.concat([f2015, f2016], axis=0)
```

```python
data
```

```python
print("*CANTIDAD de datos nulos por columna en el dataframe")
print(data.isnull().sum())
print("----------------------------------")
print("*PORCENTAJE de datos nulos por columna en el dataframe")
print(data.isnull().sum()/len(data)*100)
```

Cambia los valores nulos de las columnas "Lower Confidence Interval" y "Upper Confidence Interval" por un número aleatorio entre el valor mínimo y máximo de la misma columna (un único número, no es necesario uno diferente para cada fila con valor nulo).

```python
# Calculamos máximo de Lower Confidence Interval 
print(data.iloc[:, 12].max())
# minimo
print(data.iloc[:,12].min())
```

```python
# Calculamos máximo deUpper Confidence Interval 
print(data.iloc[:, 13].max())
# minimo
print(data.iloc[:,13].min())
```

```python
# Seleccionamos numero aleatorio entre ambos
import random
random_lower = random.uniform(2.732, 7.460)
print(random_lower)
random_upper = random.uniform(3.078, 7.669)
print(random_upper)

```

```python
#Cambiamos nan por valor aleatorio
data['Lower Confidence Interval'] = data['Lower Confidence Interval'].fillna(random_lower)
data['Upper Confidence Interval'] = data['Upper Confidence Interval'].fillna(random_upper)
```

```python
data.head(5)
```

Cambia los valores nulos de la columna "Standard Error" por su media al cuadrado.

```python
#primero obtenemos media al cuadrado de la varibale (omitiendo nan para cálculo)
sd_nonan=data['Standard Error'].dropna() 
mean2_sd=sd_nonan.mean()**2
print(mean2_sd)
```

```python
data['Standard Error'] = data['Standard Error'].fillna(mean2_sd)
data
```

```python
print("*CANTIDAD de datos nulos por columna en el dataframe")
print(data.isnull().sum())
```
